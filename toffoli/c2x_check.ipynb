```python
from qiskit import transpile
from qiskit import QuantumCircuit, QuantumRegister
from c2x_build import c2x
from c3x_build import c3x



```

```python
qc = QuantumCircuit()
q_i = QuantumRegister(3, 'q_c')
q_o = QuantumRegister(1, 'q_t')
qc.add_register(q_i)
qc.add_register(q_o)
```

```python
qc.ccx(q_i[0],q_i[1],q_o[0])
qc = qc.decompose()
qc.draw()
```

```python
c2x(qc,q_i[0],q_i[1],q_o[0])
print(qc.depth())
qc.draw()
```

```python
c3x(qc,q_i[0],q_i[1],q_i[2],q_o[0])
print(qc.depth())
qc.draw()
```
